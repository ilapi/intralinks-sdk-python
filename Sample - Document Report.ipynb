# Document Report

This report lists all the documents in the exchanges available to the end user

## Authenticate user

```python
import pandas as pd
import intralinks
import intralinks.tools.login_manager

il = intralinks.IntralinksClient()

login_manager = intralinks.tools.login_manager.LoginManager(il)

ui = intralinks.tools.login_manager.LoginManagerUI(login_manager)

ui.login_panel()
```

## Get the list of exchanges

```python
exchanges = il.get_exchanges()
```

## Get the list of documents and folders

```python
import intralinks.tools.notebooks_goodies
import intralinks.utils.associations

all_documents = []

progress_bar = intralinks.tools.notebooks_goodies.ProgressBar(items=exchanges)

for exchange in exchanges:
    progress_bar.state(message='{id} - {workspaceName}'.format(**exchange))
    
    splash = il.get_splash(exchange)
    state = il.enter_exchange(exchange, accept_splash=True)

    if state != 'ALLOW':
        # eg ONE_TIME_PASSWORD
        # eg OTP_DELIVERY_OPTION #5740955
        # eg SECURITY_QUESTION #938445
        # eg CONTENT_DECLARATION_REQUIRED #1091985 #1354425
        print('Not allowed to enter the exchange: {}'.format(state))
        continue

    folders = il.get_folders(exchange)
    documents = il.get_documents(exchange)
    intralinks.utils.associations.build_paths(folders, documents)
    
    progress_bar.state(step=1)

    for document in documents:
        all_documents.append({'document':document, 'exchange':exchange})
```

## Display the results

```python
import pandas
import pandas.io.json

dataframe = pandas.DataFrame(pandas.io.json.json_normalize(all_documents))

dataframe[[
    'document.id',
    'document.indexNumber',
    'document.name', 
    'document.extension', 
    'document.fullPath', 
    'document.fileSize',
    'document.fileHash',
    'document.createdBy.firstName', 
    'document.createdBy.lastName', 
    'document.createdOn.milliseconds', 
    'exchange.id', 
    'exchange.workspaceName'
]]
```
